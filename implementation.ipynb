```python
# Ruta de la carpeta que contiene los archivos CSV
folder_path = '/Users/simon/Documents/Protocols/EDA/CSVs'

# Obtener la lista de archivos CSV en la carpeta
csv_files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]

# DataFrame macro para almacenar los resultados de cada participante
macro_df = pd.DataFrame()

# Iterar sobre cada archivo CSV
for file in csv_files:
    # Ruta completa del archivo CSV
    csv_file_path = os.path.join(folder_path, file)
    
    # Leer el archivo CSV
    participant_df = pd.read_csv(csv_file_path)
    
    # Verificar si el archivo contiene la columna "participant"
    if 'participant' not in participant_df.columns:
        print(f"El archivo {file} no contiene la columna 'participant'. Se omitir.")
        continue
    
    # Procesar el archivo CSV utilizando la funcin preprocess_data
    participant_df = preprocess_data(csv_file_path)
    
    # Agregar el DataFrame del participante al DataFrame macro
    macro_df = pd.concat([macro_df, participant_df], ignore_index=True)

# Eliminar la columna 'img' del DataFrame macro
macro_df.drop(columns=['img'], inplace=True)

# Crear la tabla pivot con los valores sumarizados
pivot_table = macro_df.pivot_table(values=['result_count', 'react_time'], index='participant',
                                   columns=['classifier', 'result'], aggfunc={'result_count': 'sum', 'react_time': 'mean'})

# Guardar la tabla pivot en un archivo CSV
pivot_table.to_csv('/Users/simon/Documents/Protocols/EDA/macro.csv')

```
